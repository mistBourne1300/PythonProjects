## 12.15 - Exact Gradient Descent for Quadratic Functions

```python
def quad_grad_descent(Q, b, x0, eps=1e-6):
    # i = 0
	while True:
		# print(i, x0)
		# time.sleep(1)
		# i+=1
		df0 = x0.T@Q.T - b.T
		if np.linalg.norm(df0.T, ord = np.inf) < eps:
			return x0
		a0 = (df0@df0.T)/(df0@Q@df0.T)
		x1 = x0 - a0*df0.T
		x0 = x1
```

## 12.16 - Exact Gradient Descent (general)

```python
def exact_grad_descent(f, x0, eps=1e-6):
    df = elementwise_grad(f)
    i = 0
    while True:
        print(i, x0)
        i += 1
        df0 = df(x0)
        phi = lambda a: f(x0 - a*df0.T)
        alpha = newt(elementwise_grad(phi), 1., maxiter = 100000)
        x1 = x0 - alpha*df0.T
        if(np.linalg.norm(df0.T) < eps):
            return x0
        x0 = np.copy(x1)
```
